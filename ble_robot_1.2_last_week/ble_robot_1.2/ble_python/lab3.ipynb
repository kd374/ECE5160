```python
def get_data_tof2(uuid, byte_array):
    global time_IMU, data_pitch, data_roll, time_2tof, tof1_distance, tof2_distance
    
    # Convert BLE data to string
    s = ble.bytearray_to_string(byte_array).strip()

    #  DEBUG: Print raw received BLE data
    print(f"Raw received data: '{s}'")

    # Check if data is empty
    if not s:
        print("Warning: Received empty data!")
        return

    # Clean and split data
    clean_data = s.replace("|", " ").strip()
    parts = clean_data.split()

    #  DEBUG: Print split parts
    print(f"Split parts: {parts}")

    if len(parts) >= 6:
        try:
            time_IMU.append(float(parts[0]))
            data_pitch.append(float(parts[1]))
            data_roll.append(float(parts[2]))
            time_2tof.append(float(parts[3]))
            tof1_distance.append(float(parts[4]))
            tof2_distance.append(float(parts[5]))

            #  DEBUG: Print successfully appended data
            print(f"Appended Data: {time_IMU[-1]}, {data_pitch[-1]}, {data_roll[-1]}, {time_2tof[-1]}, {tof1_distance[-1]}, {tof2_distance[-1]}")
        except ValueError as e:
            print(f"Error converting to float: {e}")
    else:
        print("Unexpected format:", parts)
```
